```python
class Human:
  def __init__(self, name, age):
      self.name = name
      self.age = age

  def check_adult(self):
      if self.age >= 20:
        print("大人である")
      else:
        print("大人でない")

person1 = Human("侍太郎", 19)
person2 = Human("侍一郎", 22)
person3 = Human("侍二郎", 23)
person4 = Human("侍三郎", 18)
person5 = Human("侍四郎", 25)

list = [person1, person2, person3, person4, person5]

for person in list:
  person.check_adult()

```
